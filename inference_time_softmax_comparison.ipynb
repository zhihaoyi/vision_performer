# Inference time comparison - softmax

```python
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt
```

```python
batch_size = 64
```

```python
log_name_64 = '\log\inference_time\d_64.pkl'
log_path_64 = os.getcwd()  + log_name_64
with open(log_path_64, 'rb') as file:
    inf_time64 = pickle.load(file)
    vit_64 = inf_time64[0]
    vip_64 = inf_time64[1]
    
log_name_128 = '\log\inference_time\d_128.pkl'
log_path_128 = os.getcwd()  + log_name_128
with open(log_path_128, 'rb') as file:
    inf_time128 = pickle.load(file)
    vit_128 = inf_time128[0]
    vip_128 = inf_time128[1]
    
log_name_256 = '\log\inference_time\d_256.pkl'
log_path_256 = os.getcwd()  + log_name_256
with open(log_path_256, 'rb') as file:
    inf_time256 = pickle.load(file)
    vit_256 = inf_time256[0]
    vip_256 = inf_time256[1]
    
log_name_512 = '\log\inference_time\d_512.pkl'
log_path_512 = os.getcwd()  + log_name_512
with open(log_path_512, 'rb') as file:
    inf_time512 = pickle.load(file)
    vit_512 = inf_time512[0]
    vip_512 = inf_time512[1]
    
log_name_1024 = '\log\inference_time\d_1024.pkl'
log_path_1024 = os.getcwd()  + log_name_1024
with open(log_path_1024, 'rb') as file:
    inf_time1024 = pickle.load(file)
    vit_1024 = inf_time1024[0]
    vip_1024 = inf_time1024[1]
```

```python
logL_vit = np.log2([16, 64, 256, 1024, 4096])
logL_vip = np.log2([16, 64, 256, 1024, 4096, 16384])
plt.title(f'Inference time on {batch_size} test samples (softmax)')
plt.xlabel('log2(L)')
plt.plot(logL_vit, np.log2(vit_64), color='r', label='d=64')
plt.plot(logL_vit, np.log2(vit_128), color='g', label='d=128')
plt.plot(logL_vit, np.log2(vit_256), color='m', label='d=256')
plt.plot(logL_vit, np.log2(vit_512), color='c', label='d=512')
plt.plot(logL_vit, np.log2(vit_1024), color='y', label='d=1024')
plt.plot(logL_vip, np.log2(vip_64), '-.', color='r', label='')
plt.plot(logL_vip, np.log2(vip_128), '-.', color='g', label='')
plt.plot(logL_vip, np.log2(vip_256), '-.', color='m', label='')
plt.plot(logL_vip, np.log2(vip_512), '-.', color='c', label='')
plt.plot(logL_vip, np.log2(vip_1024), '-.', color='y', label='')
plt.ylabel('log2(Time)')

vit_handle = plt.Line2D([], [], linestyle='-')
vip_handle = plt.Line2D([], [], linestyle='-.')
red_proxy = plt.Rectangle((0, 0), 1, 1, fc='r')
green_proxy = plt.Rectangle((0, 0), 1, 1, fc='g')
m_proxy = plt.Rectangle((0, 0), 1, 1, fc='m')
c_proxy = plt.Rectangle((0, 0), 1, 1, fc='c')
yellow_proxy = plt.Rectangle((0, 0), 1, 1, fc='y')

plt.legend(handles=[red_proxy, green_proxy, m_proxy, c_proxy, yellow_proxy, vit_handle, vip_handle],
          labels=['d=64', 'd=128', 'd=256', 'd=512', 'd=1024', 'ViT', 'ViP'])
plt.grid()
```
